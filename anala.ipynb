```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("weather_data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Date"],df["Temperature"],marker="o")
plt.title("Daily Temperature Trend")
plt.xlabel("Date")
plt.ylabel("Temperature(c)")

plt.xticks(df["Date"],rotation=45)

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(df["Date"],df["Rainfall"])
plt.title("Daily Rainfall")
plt.xlabel("Date")
plt.ylabel("Rainfall (mm)")
plt.xticks(rotation=45)
plt.grid(False)
plt.show()
    
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Temperature"],bins=8)
plt.title("Temperature Distribution")
plt.xlabel("Temperature(c)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Temperature"],df["Humidity"])
plt.title("Temperature vs Humidity")
plt.xlabel("Temperature(c)")
plt.ylabel("Humidity(%)")
plt.grid(True)
plt.show()
```

```python
rain_counts=[(df["Rainfall"]==0).sum(),
             ((df["Rainfall"]>0) & (df["Rainfall"]<=5)).sum(),
             (df["Rainfall"]>5).sum()
            ]
labels=["No rain","Light rain","Heavy rain"]
plt.figure(figsize=(4,4))
plt.pie(rain_counts,labels=labels,autopct="%1.1f%%")
plt.title("Rainfall Distribution")
plt.show()
```

```python
plt.figure(figsize=(15,10))

plt.subplot(3,2,1)
plt.plot(df["Date"],df["Temperature"],marker="o")
plt.title("Temperature Trend")
plt.xticks(rotation=45)
plt.grid(True)

plt.subplot(3,2,2)
plt.bar(df["Date"],df["Rainfall"])
plt.title("Rainfall")
plt.xticks(rotation=45)

plt.subplot(3,2,3)
plt.scatter(df["Temperature"],df["Humidity"])
plt.title("Temp vs Humidity")
plt.grid(True)

plt.subplot(3,2,4)
plt.hist(df["Temperature"],bins=8)
plt.title("Temperature Distribution")
plt.grid(True)

plt.subplot(3,2,5)
plt.pie(rain_counts,labels=labels,autopct="%1.1f%%")
plt.title("Rainfall Categories")

plt.tight_layout()
plt.savefig("Temp_trends.png",dpi=300,bbox_inches="tight")
plt.show()
```

This plot shows daily temperature variation over 30 days
